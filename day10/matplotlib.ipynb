#绘制x轴和y轴的刻度

```python
from matplotlib import pyplot as plt
import random
x = range(2,26,2)
y = [random.randint(13,30) for _ in x]
plt.figure(figsize = (20,8),dpi = 100)


#plt.xticks(x)
#plt.yticks(y)
x_ticks_label = ["{}:00".format(i) for i in x]
plt.xticks(x,x_ticks_label,rotation = 45)

y_ticks_label = ["{}C".format(i) for i in range(min(y),max(y)+1)]
plt.yticks(range(min(y),max(y) + 1),y_ticks_label)
plt.plot(x,y,alpha = 0.5,color = 'yellow')
plt.show()
```

#设置显示中文

```python
from matplotlib import pyplot as plt
from matplotlib import font_manager
import random
import matplotlib
x = range(0,120)
y = [random.randint(10,50) for _ in x]
plt.figure(figsize = (20,8),dpi = 100)

my_font = font_manager.FontProperties(fname = "C:\\Windows\\Fonts\\simsun.ttc",size = 18)
plt.xlabel("时间",rotation = 45,font_properties = my_font)
plt.ylabel("次数",font_properties = my_font)
plt.title("每分钟跳动次数",color = 'red',font_properties = my_font)
plt.plot(x,y)
plt.show()
```
